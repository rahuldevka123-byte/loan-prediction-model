```python
#clearing the null  values
data['LoanAmount'] = data['LoanAmount'].fillna(data['LoanAmount'].mean())
```

```python
#clearing the null  values
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].median())
```

```python
#Displaying the items which has null values
data.isnull().sum()
```

```python
#Cleared the null values  using mean, median , mode
data['LoanAmount'] = data['LoanAmount'].fillna(data['LoanAmount'].mean())
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].median())
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0])
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
data['Married'] = data['Married'].fillna(data['Married'].mode()[0])
data['Dependents'] = data['Dependents'].fillna(data['Dependents'].mode()[0])
data['Self_Employed'] = data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
```

```python
#Rechecking after clearing all null
data.isnull().sum()
```

```python
data['Loan_Status'].replace('Y',1,inplace=True)
data['Loan_Status'].replace('N',0,inplace=True)
```

```python
data['Loan_Status'].value_counts()
```

```python
data.Gender=data.Gender.map({'Male':1,'Female':0})
data['Gender'].value_counts()
```

```python
data.Married=data.Married.map({'Yes':1,'No':0})
data['Married'].value_counts()
```

```python
data.Dependents=data.Dependents.map({'0':0,'1':1,'2':2,'3+':3})
data['Dependents'].value_counts()
```

```python
data.Education=data.Education.map({'Graduate':1,'Not Graduate':0})
data['Education'].value_counts()
```

```python

data.Self_Employed=data.Self_Employed.map({'Yes':1,'No':0})
data['Self_Employed'].value_counts()
```

```python
data.Property_Area=data.Property_Area.map({'Urban':2,'Rural':0,'Semiurban':1})
data['Property_Area'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
```

```python
X = data.iloc[1:542,1:12].values
y = data.iloc[1:542,12].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=0)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("✅ Model Accuracy:", acc)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Precision
prec = precision_score(y_test, y_pred)
print("\nPrecision:", prec)

# Recall
rec = recall_score(y_test, y_pred)
print("Recall:", rec)
```

```python
# ------------------ USER INPUT ------------------




# Take inputs
Gender = int(input("Gender (1=Male, 0=Female): "))
Married = int(input("Married (1=Yes, 0=No): "))
Dependents = int(input("Dependents (0,1,2,3): "))
Education = int(input("Education (1=Graduate, 0=Not Graduate): "))
Self_Employed = int(input("Self Employed (1=Yes, 0=No): "))
ApplicantIncome = int(input("Applicant Income (monthly in ₹): "))
CoapplicantIncome = float(input("Coapplicant Income (monthly in ₹): "))
LoanAmount = float(input("Loan Amount (in thousands, e.g., 200 = ₹2,00,000): "))
Loan_Amount_Term = float(input("Loan Amount Term (in months, e.g., 360 = 30 years): "))

# Loan type
print("\nSelect Loan Type:")
print("1 = Home Loan")
print("2 = Car Loan")
print("3 = Personal Loan")
print("4 = Education Loan")
loan_type = int(input("Enter loan type (1-4): "))

# CIBIL and existing loans
CIBIL_Score = int(input("CIBIL Score (300-900): "))
existing_loans = int(input("Number of active loans you currently have: "))

# Property area
Property_Area = int(input("Property Area (0=Rural, 1=Semiurban, 2=Urban): "))


# Set minimum CIBIL based on loan type
if loan_type == 1:
    min_cibil = 700  # Home Loan
elif loan_type == 2:
    min_cibil = 650  # Car Loan
elif loan_type == 3:
    min_cibil = 650  # Personal Loan
elif loan_type == 4:
    min_cibil = 600  # Education Loan
else:
    min_cibil = 700  # default

# Convert CIBIL and existing loans to Credit_History
if CIBIL_Score >= min_cibil and existing_loans < 2:
    Credit_History = 1  # Good
else:
    Credit_History = 0  # Bad / risky

# Optional warning for multiple loans
if existing_loans >= 2:
    print("\n You already have multiple loans; approval may be affected.")

# ------------------ PREDICTION ------------------
user_data = [[Gender, Married, Dependents, Education, Self_Employed,
              ApplicantIncome, CoapplicantIncome, LoanAmount,
              Loan_Amount_Term, Credit_History, Property_Area]]

prediction = model.predict(user_data)

# Show result
if prediction[0] == 1:
    print("\n Congratulations! Loan will be APPROVED.")
else:
    print("\n Sorry, Loan will be REJECTED.")
```
